```python
from bs4 import BeautifulSoup
import requests

url ='https://wuzzuf.net'
web = f'{url}/search/jobs/?q=data%20anaysis%20&a=hpb'
response = requests.get(web)
content = response.text
soup = BeautifulSoup(content , 'lxml')

box=soup.find('div', class_= "css-9i2afk")
container= soup.find_all('div', class_="css-1gatmva e1v1l3u10")

t=[]
c=[]
l=[]
j=[]
s=[]

pagination = soup.find('div' , class_ = 'css-7o92qm')
pages = pagination.find_all('li' , class_ = 'css-1q4vxyr')
last_page = pages[-2].text

for page in range (0, int(last_page)+1)[:3]:
    website = 'https://wuzzuf.net/search/jobs/?a=hpb&q=data%20anaysis%20&start'
    response= requests.get(f'{website}/{page}')
    content= response.text
    soup = BeautifulSoup(content , 'lxml')
    container= soup.find_all('div', class_="css-1gatmva e1v1l3u10")
    
    for cont in container:
        t.append(cont.find('h2' , class_= "css-m604qf").get_text().strip())
        c.append(cont.find('a' , class_= "css-17s97q8").get_text().strip())
        l.append(cont.find('span' , class_= "css-5wys0k").get_text().strip())
        j.append(cont.find('span' , class_= "css-1ve4b75 eoyjyou0").get_text().strip())
        s.append(cont.find('span' , class_= "css-o1vzmt eoyjyou0").get_text().strip())
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({"Title":t , 'company_name':c , 'Location':l , 'Type':j , 'Site':s})
df
```
